```python
# Increase Brightness
bright = cv2.convertScaleAbs(image, alpha=1, beta=50)

# Decrease Brightness
dark = cv2.convertScaleAbs(image, alpha=1, beta=-50)

# Display Results
show("Original Image", image)
show("Increased Brightness", bright)
show("Decreased Brightness", dark)
```

```python
# Increase Contrast
high_contrast = cv2.convertScaleAbs(image, alpha=1.8, beta=0)

# Decrease Contrast
low_contrast = cv2.convertScaleAbs(image, alpha=0.6, beta=0)

# Display Results
show("Original Image", image)
show("High Contrast", high_contrast)
show("Low Contrast", low_contrast)
```

```python
# Apply Gaussian Blur

gaussian = cv2.GaussianBlur(image, (15, 15), 0)

# Display Results
show("Original Image", image)
show("Gaussian Blur", gaussian)
```

```python
# Apply Median Blur

median = cv2.medianBlur(image, 9)

# Display Results
show("Original Image", image)
show("Median Blur", median)
```

```python
# Apply Bilateral Filter

bilateral = cv2.bilateralFilter(image, 9, 75, 75)

# Display Results
show("Original Image", image)
show("Bilateral Filter", bilateral)
```

```python
# Sharpening Kernel
kernel = np.array([
    [-1, -1, -1],
    [-1,  9, -1],
    [-1, -1, -1]
])

# Apply Sharpening
sharpened = cv2.filter2D(image, -1, kernel)

# Display Results
show("Original Image", image)
show("Sharpened Image", sharpened)
```
