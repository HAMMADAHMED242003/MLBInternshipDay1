```python
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras import layers, models
```

```python
base_model = MobileNetV2(
    weights='imagenet',
    include_top=False,
    input_shape=(224, 224, 3)
)
```

```python
base_model.summary()
```

```python
base_model.trainable = False
```

```python
print(base_model.trainable)
```

```python
model = models.Sequential([
    base_model,

    layers.GlobalAveragePooling2D(),

    layers.Dense(128, activation='relu'),

    layers.Dropout(0.2),

    layers.Dense(1, activation='sigmoid')
])
```

```python
model.summary()
```

```python
import tensorflow as tf
import tensorflow_datasets as tfds
```

```python
(train_ds, validation_ds), dataset_info = tfds.load(
    'cats_vs_dogs',
    split=['train[:80%]', 'train[80%:]'],
    as_supervised=True,
    with_info=True
)
```

```python
print(dataset_info.features)
```

```python
for image, label in train_ds.take(1):
    print(image.shape)
    print(label)
```

```python
IMG_SIZE = (224, 224)

def preprocess(image, label):
    image = tf.image.resize(image, IMG_SIZE)
    image = image / 255.0
    return image, label
```

```python
train_ds = train_ds.map(preprocess)
validation_ds = validation_ds.map(preprocess)
```

```python
BATCH_SIZE = 32

train_ds = train_ds.batch(BATCH_SIZE)
validation_ds = validation_ds.batch(BATCH_SIZE)
```

```python
train_ds = train_ds.prefetch(tf.data.AUTOTUNE)
validation_ds = validation_ds.prefetch(tf.data.AUTOTUNE)
```

```python
for images, labels in train_ds.take(1):
    print(images.shape)
    print(labels.shape)
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
history = model.fit(
    train_ds,
    validation_data=validation_ds,
    epochs=5
)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.plot(history.history['accuracy'], marker='o', label='Training Accuracy')
plt.plot(history.history['val_accuracy'], marker='o', label='Validation Accuracy')

plt.title("Training vs Validation Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.savefig("accuracy_graph.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.plot(history.history['loss'], marker='o', label='Training Loss')
plt.plot(history.history['val_loss'], marker='o', label='Validation Loss')

plt.title("Training vs Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.savefig("loss_graph.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
import matplotlib.pyplot as plt
import tensorflow as tf

class_names = ["Cat", "Dog"]

# Shuffle the validation dataset
random_ds = validation_ds.unbatch().shuffle(1000).batch(9)

for images, labels in random_ds.take(1):

    predictions = model.predict(images)

    plt.figure(figsize=(12,12))

    for i in range(9):

        plt.subplot(3,3,i+1)

        plt.imshow(images[i])

        pred = 1 if predictions[i] > 0.5 else 0

        plt.title(
            f"Pred: {class_names[pred]}\nActual: {class_names[labels[i]]}"
        )

        plt.axis("off")

    plt.tight_layout()
    plt.savefig("Samples.png", dpi=300, bbox_inches="tight")
    plt.show()
```
