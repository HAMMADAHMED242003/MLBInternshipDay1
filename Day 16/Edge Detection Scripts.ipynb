```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("/kaggle/input/datasets/sp23bse071/coinimage/Coins.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(6,6))
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

plt.figure(figsize=(6,6))
plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

```python
blur = cv2.GaussianBlur(gray, (15,15), 0)

plt.figure(figsize=(6,6))
plt.imshow(blur, cmap='gray')
plt.title("Gaussian Blur")
plt.axis("off")
plt.show()
```

```python
sobelx = cv2.Sobel(blur, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(blur, cv2.CV_64F, 0, 1, ksize=3)

sobel = cv2.magnitude(sobelx, sobely)

plt.figure(figsize=(6,6))
plt.imshow(sobel, cmap='gray')
plt.title("Sobel Edge Detection")
plt.axis("off")
plt.show()
```

```python
laplacian = cv2.Laplacian(blur, cv2.CV_64F)

plt.figure(figsize=(6,6))
plt.imshow(laplacian, cmap='gray')
plt.title("Laplacian Edge Detection")
plt.axis("off")
plt.show()
```

```python
canny = cv2.Canny(blur,100,200)

plt.figure(figsize=(6,6))
plt.imshow(canny, cmap='gray')
plt.title("Canny Edge Detection")
plt.axis("off")
plt.show()
```

```python
fig, ax = plt.subplots(1,3, figsize=(18,6))

ax[0].imshow(sobel,cmap='gray')
ax[0].set_title("Sobel")

ax[1].imshow(laplacian,cmap='gray')
ax[1].set_title("Laplacian")

ax[2].imshow(canny,cmap='gray')
ax[2].set_title("Canny")

for a in ax:
    a.axis("off")

plt.show()
```

```python
# Convert grayscale image into binary image
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV)
plt.figure(figsize=(6,6))
plt.imshow(binary, cmap='gray')
plt.title("Binary Image")
plt.axis("off")
plt.show()
```

```python
kernel = np.ones((5,5), np.uint8)
```

```python
erosion = cv2.erode(binary, kernel, iterations=1)

plt.figure(figsize=(6,6))
plt.imshow(erosion, cmap='gray')
plt.title("Erosion")
plt.axis("off")
plt.show()
```

```python
dilation = cv2.dilate(binary, kernel, iterations=1)

plt.figure(figsize=(6,6))
plt.imshow(dilation, cmap='gray')
plt.title("Dilation")
plt.axis("off")
plt.show()
```

```python
opening = cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel)

plt.figure(figsize=(6,6))
plt.imshow(opening, cmap='gray')
plt.title("Opening")
plt.axis("off")
plt.show()
```

```python
closing = cv2.morphologyEx(binary, cv2.MORPH_CLOSE, kernel)

plt.figure(figsize=(6,6))
plt.imshow(closing, cmap='gray')
plt.title("Closing")
plt.axis("off")
plt.show()
```

```python
gradient = cv2.morphologyEx(binary, cv2.MORPH_GRADIENT, kernel)

plt.figure(figsize=(6,6))
plt.imshow(gradient, cmap='gray')
plt.title("Morphological Gradient")
plt.axis("off")
plt.show()
```

```python
tophat = cv2.morphologyEx(binary, cv2.MORPH_TOPHAT, kernel)

plt.figure(figsize=(6,6))
plt.imshow(tophat, cmap='gray')
plt.title("Top Hat")
plt.axis("off")
plt.show()
```

```python
blackhat = cv2.morphologyEx(binary, cv2.MORPH_BLACKHAT, kernel)

plt.figure(figsize=(6,6))
plt.imshow(blackhat, cmap='gray')
plt.title("Black Hat")
plt.axis("off")
plt.show()
```

```python
titles = [
    "Binary",
    "Erosion",
    "Dilation",
    "Opening",
    "Closing",
    "Gradient",
    "Top Hat",
    "Black Hat"
]

images = [
    binary,
    erosion,
    dilation,
    opening,
    closing,
    gradient,
    tophat,
    blackhat
]

plt.figure(figsize=(18,10))

for i in range(len(images)):
    plt.subplot(2,4,i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis("off")

plt.tight_layout()
plt.show()
```
