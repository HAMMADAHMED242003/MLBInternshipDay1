```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
iris=load_iris()
```

```python
print(iris.keys())
```

```python
df= pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
plt.scatter(df["sepal length (cm)"], df["sepal width (cm)"])
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Original Iris Data")
plt.savefig("original_iris_data.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
```

```python
kmeans.fit(df)
```

```python
clusters = kmeans.labels_
```

```python
clusters = kmeans.fit_predict(df)
```

```python
print(clusters)
```

```python
df["Cluster"] = clusters
```

```python
df.head(70)
```

```python
df.columns
```

```python
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(df.iloc[:, :4])
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), wcss, marker='o')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.savefig("Elbow.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
plt.scatter(
    df.iloc[:,0],
    df.iloc[:,1],
    c=df["Cluster"]
)

plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("K-Means Clusters")
plt.savefig("K-Means Cluser.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
pca = PCA(n_components=2)
```

```python
pca_data = pca.fit_transform(df.iloc[:, :4])
```

```python
print(pca_data)
```

```python
pca_df = pd.DataFrame(
    pca_data,
    columns=["PC1", "PC2"]
)
```

```python
pca_df.head()
```

```python
plt.scatter(
    pca_df["PC1"],
    pca_df["PC2"]
)

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA Visualization")
plt.savefig("PCA Visualization.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
pca_df["Cluster"] = df["Cluster"]
```

```python
plt.scatter(
    pca_df["PC1"],
    pca_df["PC2"],
    c=pca_df["Cluster"]
)

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA with K-Means Clusters")
plt.savefig("PCA with K-Means Clusters", dpi=300, bbox_inches="tight")
plt.show()
```
