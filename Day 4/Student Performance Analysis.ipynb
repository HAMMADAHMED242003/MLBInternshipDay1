```python
mathAvg=int(df['math score'].mean())
readingAvg=int(df['reading score'].mean())
writingAvg=int(df['writing score'].mean())

print("Math Score average is: ",mathAvg)
print("Math Score average is: ",readingAvg)
print("Math Score average is: ",writingAvg)
```

```python
#Making new Column
df['TotalMarks']= df['math score'] + df['reading score'] + df['writing score']
```

```python
print(df.columns)
```

```python
top_5 = df.nlargest(5, 'TotalMarks')
print(top_5)
```

```python
#Making new Column
df['Average']= df['math score'].mean() + df['reading score'].mean() + df['writing score'].mean()
```

```python
df[df['TotalMarks']<df['Average']]
```
