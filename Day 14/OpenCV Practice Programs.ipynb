```python
import cv2
import os
import numpy as np
```

```python
image_path = "/kaggle/input/datasets/sp23bse071/catimage/cat.jpg"
```

```python
image = cv2.imread(image_path)
```

```python
if image is None:
    print("Image not found!")
else:
    print("Image loaded successfully.")
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

```python
height, width, channels = image.shape

print("Height :", height)
print("Width :", width)
print("Channels :", channels)
```

```python
file_size = os.path.getsize(image_path)
```

```python
file_size = file_size / 1024

print(f"File Size : {file_size:.2f} KB")
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(6,6))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.imshow(gray, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

```python
print("Original Shape:", image.shape)
print("Grayscale Shape:", gray.shape)
```

```python
small = cv2.resize(image, (300, 200))
medium = cv2.resize(image, (600, 400))
large = cv2.resize(image, (900, 600))
```

```python
print("Original:", image.shape)
print("Small:", small.shape)
print("Medium:", medium.shape)
print("Large:", large.shape)
```

```python
plt.figure(figsize=(5,5))
plt.imshow(cv2.cvtColor(small, cv2.COLOR_BGR2RGB))
plt.title("Small")
plt.axis("off")
plt.show()

plt.figure(figsize=(5,5))
plt.imshow(cv2.cvtColor(medium, cv2.COLOR_BGR2RGB))
plt.title("Medium")
plt.axis("off")
plt.show()

plt.figure(figsize=(5,5))
plt.imshow(cv2.cvtColor(large, cv2.COLOR_BGR2RGB))
plt.title("Large")
plt.axis("off")
plt.show()
```

```python
crop1 = image[0:300, 0:300]
```

```python
crop2 = image[150:500, 250:650]
```

```python
crop3 = image[300:700, 500:900]
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15,5))

axes[0].imshow(cv2.cvtColor(crop1, cv2.COLOR_BGR2RGB))
axes[0].set_title("Crop 1")
axes[0].axis("off")

axes[1].imshow(cv2.cvtColor(crop2, cv2.COLOR_BGR2RGB))
axes[1].set_title("Crop 2")
axes[1].axis("off")

axes[2].imshow(cv2.cvtColor(crop3, cv2.COLOR_BGR2RGB))
axes[2].set_title("Crop 3")
axes[2].axis("off")

plt.show()
```

```python
import os

os.makedirs("output", exist_ok=True)

cv2.imwrite("output/crop1.jpg", crop1)
cv2.imwrite("output/crop2.jpg", crop2)
cv2.imwrite("output/crop3.jpg", crop3)
```

```python
rotate90 = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
```

```python
rotate180 = cv2.rotate(image, cv2.ROTATE_180)
```

```python
rotate270 = cv2.rotate(image, cv2.ROTATE_90_COUNTERCLOCKWISE)
```

```python
print("Original :", image.shape)
print("90° :", rotate90.shape)
print("180° :", rotate180.shape)
print("270° :", rotate270.shape)
```

```python
fig, axes = plt.subplots(1, 4, figsize=(20,5))

axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[0].set_title("Original")
axes[0].axis("off")

axes[1].imshow(cv2.cvtColor(rotate90, cv2.COLOR_BGR2RGB))
axes[1].set_title("90°")
axes[1].axis("off")

axes[2].imshow(cv2.cvtColor(rotate180, cv2.COLOR_BGR2RGB))
axes[2].set_title("180°")
axes[2].axis("off")

axes[3].imshow(cv2.cvtColor(rotate270, cv2.COLOR_BGR2RGB))
axes[3].set_title("270°")
axes[3].axis("off")

plt.show()
```

```python

cv2.imwrite("output/rotate90.jpg", rotate90)
cv2.imwrite("output/rotate180.jpg", rotate180)
cv2.imwrite("output/rotate270.jpg", rotate270)

print("Images saved successfully!")
```

```python
horizontal = cv2.flip(image, 1)
```

```python
vertical = cv2.flip(image, 0)
```

```python
both = cv2.flip(image, -1)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15,5))

axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[0].set_title("Original")
axes[0].axis("off")

axes[1].imshow(cv2.cvtColor(horizontal, cv2.COLOR_BGR2RGB))
axes[1].set_title("Horizontal Flip")
axes[1].axis("off")

axes[2].imshow(cv2.cvtColor(vertical, cv2.COLOR_BGR2RGB))
axes[2].set_title("Vertical Flip")
axes[2].axis("off")

plt.show()
```

```python

cv2.imwrite("output/horizontal_flip.jpg", horizontal)
cv2.imwrite("output/vertical_flip.jpg", vertical)

print("Flipped images saved successfully!")
```

```python
cv2.rectangle(
    image,
    (50, 50),       # Top-left corner
    (250, 250),     # Bottom-right corner
    (0, 255, 0),    # Green (BGR)
    3               # Thickness
)
```

```python
cv2.circle(
    image,
    (450, 150),     # Center
    60,             # Radius
    (255, 0, 0),    # Blue
    3
)
```

```python
cv2.line(image, (100, 200), (500, 200), (0, 0, 255), 3)
```

```python
points = np.array([
    [420,100],
    [520,150],
    [500,250],
    [400,250],
    [380,150]
], np.int32)

cv2.polylines(image, [points], True, (255,0,255), 3)
```
